```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# distribution of legit and fraud transactions
```

```python
credit_card['Class'].value_counts()  # highly unbalanced  dataset
```

```python
# seprate the data based on class for analysis
legit = credit_card[credit_card['Class'] == 0]
fraud = credit_card[credit_card['Class'] == 1]
```

```python
credit_card.groupby('Class').mean()
```

```python
# build a sample dataset having similar distribution of normal 
# transactions and fraudlent transactions
```

```python
legit_sample = legit.sample(n=492)
```

```python
len(legit_sample)
```

```python
df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
df.head(3)
```

```python
df['Class'].value_counts()
```

```python
df.groupby('Class').mean()
```

```python
# splitting into the data into features and targets
```

```python
x = df.iloc[:,0:30]
y = df['Class']
```

```python
x.head()
```

```python
y
```

```python
# split the data into test and train
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 ,stratify = y, random_state = 42)
```

```python
x.shape , x_train.shape , x_test.shape
```

```python
# train the model 
```

```python
model = LogisticRegression()
model.fit(x_train , y_train)
```

```python
# model evalution
```

```python
# accuracy on train data
```

```python
y_pred_train = model.predict(x_train)
accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_train
```

```python
# accuracy on test data
y_pred_test = model.predict(x_test)
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_test
```
